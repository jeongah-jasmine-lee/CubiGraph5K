```python
import re
import shutil
from pathlib import Path

# --- Configuration ---
result_file = Path("./resampling-result.txt")
base_src     = Path("../cubicasa5k-3493-cubigraph")
dest_root    = Path("../cubicasa5k-228")

# --- Step 1: Extract source directories from the text file ---
pattern = re.compile(r"(\.\./cubicasa5k-3493-cubigraph/[^\s]+)")
src_paths = []

with result_file.open() as f:
    for line in f:
        m = pattern.search(line)
        if m:
            src_paths.append(Path(m.group(1)))

# --- Step 2: Create destination root ---
dest_root.mkdir(parents=True, exist_ok=True)

# --- Step 3: Copy each folder ---
for src in src_paths:
    # Make the subpath relative to the base source root
    rel = src.relative_to(base_src)
    dest = dest_root / rel
    
    # Ensure parent directories exist
    dest.parent.mkdir(parents=True, exist_ok=True)
    
    # Copy entire directory tree
    print(f"Copying {src} → {dest}")
    shutil.copytree(src, dest)

print("All done! Copied", len(src_paths), "folders into", dest_root)
```
