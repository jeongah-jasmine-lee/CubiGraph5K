```python
import os
import shutil
from pathlib import Path

# 경로 세팅
result_txt_path = "./resampling-result.txt"  # 텍스트 파일 경로
src_root = "../cubicasa5k-666"
dst_root = "../cubicasa5k-114"

os.makedirs(dst_root, exist_ok=True)

# 1. txt 파일에서 Easy/Medium 경로 파싱
selected_dirs = []

with open(result_txt_path, 'r') as f:
    for line in f:
        line = line.strip()
        if line.startswith("[Easy]") or line.startswith("[Medium]"):
            # 예: "[Easy] ../cubicasa5k-666/high_quality/641 (score: 0.2099)"
            path_part = line.split(']')[1].strip().split(" (score")[0].strip()
            selected_dirs.append(path_part)

# 2. 복사 수행
copied = 0
for src_dir in selected_dirs:
    src_dir = os.path.normpath(src_dir)
    src_dir_abs = os.path.abspath(src_dir)

    # 상대 경로 추출
    rel_path = os.path.relpath(src_dir, start=src_root)
    dst_dir = os.path.join(dst_root, rel_path)

    # 복사
    if os.path.exists(src_dir_abs):
        shutil.copytree(src_dir_abs, dst_dir, dirs_exist_ok=True)
        copied += 1
        print(f"Copied from {src_dir_abs} to {dst_dir}")
    else:
        print(f"[❗경고] 원본 폴더 없음: {src_dir_abs}")

print(f"\n✅ 총 {copied}개의 폴더가 '{dst_root}'에 복사 완료되었습니다.")
```
